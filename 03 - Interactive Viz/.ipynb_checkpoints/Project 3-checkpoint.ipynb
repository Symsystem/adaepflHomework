```python
import folium
import pandas as pd

cantons = 'ch-cantons.topojson.json'
grant_export = 'P3_GrantExport.csv'
data  = pd.read_csv(grant_export, sep=';', error_bad_lines=False)


print(data.shape)
```

```python
L = []
for st in data['University'] :
    a=str(st).split(" - ")
    if len(a) > 1 :
        L.append( str(st).split(" - ")[1])
    else :
        L.append('NaN')
data['Shortcode Uni'] = L
```

```python
count_university = data['Shortcode Uni'].value_counts()
```

```python
df = pd.DataFrame(data=count_university)
df.head(100)
```

```python
#Let Folium determine the scale
map = folium.Map(location=[47, 8], zoom_start=3)
#map.choropleth(geo_path =cantons ,topojson=cantons)


map.choropleth(geo_path=cantons, data=df,
              key_on='feature.id',
              fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2,
              legend_name='Test')

#you have to save the map in an htlm ...
map.save('test.html')
```
