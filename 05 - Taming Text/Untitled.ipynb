```python
# Tokenization \n ,
import string
from nltk import word_tokenize
text_tokened = [i for i in word_tokenize(text.lower())]

stop = set(stopwords.words('english'))
stop.extend(["n't", 'pm'])
stop.update(string.punctuation) #Remove ponctuation
body_updated_tokenized = [i for i in word_tokenize(text.lower()) if i not in stop]
```
